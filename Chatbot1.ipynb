## Chatbot with using the tkinter 

```python
import pyttsx3
import random
import tkinter as tk
from tkinter import scrolledtext

# Chatbot logic
def greeting():
    greetings = ["Hello there! How can I help you?"]                   ## make the fuction to wishing first
    return random.choice(greetings)

def farewell():
    farewells = ["Goodbye!"]                                           ## make the fuction to end greeting
    return random.choice(farewells)

def basic_questions():
    questions = [
        "What is your name?",
        "Where are you from?",                                      ## fuction that  store the five question 
        "What is your favorite hobby?",
        "Do you have any pets?",
        "What is your favorite color?"
    ]
    return random.choice(questions)

def respond(question):
    def respond(question,previous_contexts):
        if question == "What is your name?":
            return "My name is Chatbot!"
        elif question == "Where are you from?":
            return "I don't have a physical location"                  ## this fuction is respond the anser which i asked from out of 5 question
        elif question == "What is your favourite hooby?":
            return " learning the new things"
        elif question == "Who can built you?":
            return "I don't know"
        elif question == "What is your favourite color?":
            return "I have no favourite color"
        else:
            return "I'm not sure to respond to that.Please try asking a different question."





                                   # Front-end using Tkinter


def chat():
    def send():
        user_input = entry_field.get()
        if user_input == "":                                                                              # Check if the input is empty
            thank_you_message = "Thank you for using the chatbot!GoodBye! Have a nice day!"
            output_box.insert(tk.END, "Bot: " + thank_you_message + "\n")
            engine.say(thank_you_message)
            engine.runAndWait()
            return

        
        if user_input.lower() == "quit":
            output_box.insert(tk.END, "Bot: " + farewell() + "\n")
            engine.say(farewell())
            engine.runAndWait()
            root.quit()
            return
        
        output_box.insert(tk.END, "You: " + user_input + "\n")
        response = respond(user_input)
        output_box.insert(tk.END, "Bot: " + response + "\n")

        engine.say(response)
        engine.runAndWait()

        entry_field.delete(0, tk.END)  # Clear the input field

                                                                                 # Text-to-Speech engine
    engine = pyttsx3.init()

                                                               # Tkinter window setup
    root = tk.Tk()
    root.title("Chatbot")

                                                                                       # Scrollable output box
    output_box = scrolledtext.ScrolledText(root, wrap=tk.WORD, width=50, height=15, state='normal')
    output_box.pack(pady=10)

    # User input field
    entry_field = tk.Entry(root, width=50)
    entry_field.pack(pady=5)

 
    send_button = tk.Button(root, text="Send", command=send)
    send_button.pack(pady=5)


    output_box.insert(tk.END, "Bot: " + greeting() + "\n")
    engine.say(greeting())
    engine.runAndWait()

  
    root.mainloop()

if __name__ == "__main__":
    chat()
```
